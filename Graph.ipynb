```python
myfile = open("Q1b.out")
data = []
for line in myfile:
    data.append(line)
myfile.close()
```

```python
time = []
core = []
for val in data:
    val = val.split()
    time.append(float(val[0]))
    core.append(float(val[1]))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(core, time)
plt.title("Simulation Time on Using Different Cores")
plt.xlabel("Number of Cores")
plt.ylabel("Seconds")
plt.savefig("Q1b.png")
```

The speed up is not linear because the running time of using 1 core is already very low, there is not much potential to decrease the running time. In addition to that, running on mutiple cores require inter-connection between each core to share their results, which could delay the whole process.

```python
myfile = open("Q2b.out")
data = []
for line in myfile:
    data.append(line)
myfile.close()
```

```python
period = []
rhos = []
for i, row in enumerate(data):
    if (i == 50) | (i == 0):
        ts = row.split()[1:]
        period.append(float(ts[0]))
        period.append(float(ts[2]))
        rhos.append(float(ts[1]))
        rhos.append(float(ts[3][1:]))
    elif i == 49:
        ts = row.split()
        period.append(float(ts[0]))
        period.append(float(ts[2]))
        rhos.append(float(ts[1]))
        rhos.append(float(ts[3][:-1]))
    elif i == 99:
        ts = row.split()
        period.append(float(ts[0]))
        period.append(float(ts[2]))
        rhos.append(float(ts[1]))
        rhos.append(float(ts[3][:-2]))
    else:
        ts = row.split()
        period.append(float(ts[0]))
        period.append(float(ts[2]))
        rhos.append(float(ts[1]))
        rhos.append(float(ts[3]))
```

```python
plt.scatter(rhos, period)
plt.title("Living Period on Using Different Rhos")
plt.xlabel("Rhos")
plt.ylabel("Periods")
plt.savefig("Q2b.png")
```

The computation time I get is 0.8806726932525635 seconds, the optimal rho is 0.004773869346733628 and the average living period is 748.476
